## Utilizando astroquery
Más tablas en http://cdsxmatch.u-strasbg.fr/xmatch/doc/available-tables.html

```python
from astropy import units as u
from astroquery.xmatch import XMatch
table = XMatch.query(cat1=data['RA','DEC'],
                    cat2='vizier:I/339/hsoy',
                    max_distance=5 * u.arcsec, colRA1='RA',
                    colDec1='DEC')
```

## Utilizando VOSA
Converitr a VOTable para subir a la página
